```python
import re
```

```python
phoneRegex = re.compile(r'\d\d\d-\d\d\d-\d\d\d\d')
mos = phoneRegex.search('My number is 123-456-7890') # matched objects
print(mos.group())
```

```python
phoneRegex = re.compile(r'(\d\d\d)-(\d\d\d-\d\d\d\d)')
mos = phoneRegex.search('My number is 123-456-7890')
print(mos.group(1))
print(mos.group(2))
```

```python
phoneRegex = re.compile(r'\(\d\d\d\)-\d\d\d-\d\d\d\d')
mos = phoneRegex.search('My number is (123)-456-7890')
print(mos.group())
```

```python
batRegex = re.compile(r'Bat(wo)?man')

mos = batRegex.search('Adventure of Batman')
print(mos.group())

mos = batRegex.search('Adventure of Batwoman')
print(mos.group())

mos = batRegex.search('Adventure of Batwowowowoman')
print(mos == None)
```

```python
batRegex = re.compile(r'Bat(wo)*man')

mos = batRegex.search('Adventure of Batman')
print(mos.group())

mos = batRegex.search('Adventure of Batwoman')
print(mos.group())

mos = batRegex.search('Adventure of Batwowowowoman')
print(mos.group())
```

```python
batRegex = re.compile(r'Bat(wo)+man')

mos = batRegex.search('Adventure of Batman')
print(mos == None)

mos = batRegex.search('Adventure of Batwoman')
print(mos.group())

mos = batRegex.search('Adventure of Batwowowowoman')
print(mos.group())
```

```python
batRegex = re.compile(r'(Ha){3}')

mos = batRegex.search('Ha')
print(mos == None)

mos = batRegex.search('HaHa')
print(mos == None)

mos = batRegex.search('HaHaHa')
print(mos.group())

mos = batRegex.search('HaHaHaHa')
print(mos == None)
```

```python
batRegex = re.compile(r'(Ha){3,5}')

mos = batRegex.search('HaHa')
print(mos == None)

mos = batRegex.search('HaHaHa')
print(mos.group())

mos = batRegex.search('HaHaHaHa')
print(mos.group())

mos = batRegex.search('HaHaHaHaHa')
print(mos == None)

mos = batRegex.search('HaHaHaHaHaHa')
print(mos == None)
```

```python
text = '<to serve humans> for dinner>'
regex = re.compile(r'<(.*)>') # greedy
print(regex.findall(text))

regex = re.compile(r'<(.*?)>') # non greedy
print(regex.findall(text))
```

```python
text = '<to serve humans> for dinner\nthis evening>'
regex = re.compile(r'<(.*)>') # greedy
print(regex.findall(text))

regex = re.compile(r'<(.*?)>') # non greedy
print(regex.findall(text))

regex = re.compile(r'<(.*)>', re.DOTALL) # greedy
print(regex.findall(text))

regex = re.compile(r'<(.*?)>', re.DOTALL) # greedy
print(regex.findall(text))
```

```python
text = '<To sErvE humans> for dinner\nThis evening>'

regex = re.compile(r'[aeiou]') # greedy
print(regex.findall(text))

regex = re.compile(r'[aeiou]', re.IGNORECASE) # greedy
print(regex.findall(text))

regex = re.compile(r'[aeiou]', re.I) # greedy
print(regex.findall(text))
```

```python
import re
import pyperclip

phoneRegex = re.compile(r'''
# 123-456-7890, 456-7890, (123) 456-7890, 456-7890 ext 12345, ext. 12345, x12345
(
((\d\d\d)|(\(\d\d\d\)))?      # area code
(\s|-)             # first separator
\d\d\d             # first 3 digits
-                  # separator
\d\d\d\d           # last 4 digits
(((ext(\.)?\s)|x)   # extension word-part
(\d{2, 5}))?        # extension numeric part
)
''', re.VERBOSE)

emailRegex = re.compile(r'''
# some.+_thing@something.com

[a-zA-Z0-9_.+]+   # name part
@                 # @ synbol
[a-zA-Z0-9_.+]+   # domain name
''', re.VERBOSE)

text = pyperclip.paste()

extractedPhone = phoneRegex.findall(text)
extractedEmail = emailRegex.findall(text)
allPhones = [x[0] for x in extractedPhone]

#for i, j in zip(allPhones, extractedEmail):
#    print(i, ': ', j)

phoneJoined = '\n'.join(allPhones)
emailJoined = '\n'.join(extractedEmail)
result = phoneJoined + emailJoined
print(result)
```
